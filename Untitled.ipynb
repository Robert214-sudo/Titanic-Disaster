```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load the data
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

# Save test PassengerId for submission
passenger_ids = test_df['PassengerId']

# Combine train and test for consistent preprocessing
train_df['TrainFlag'] = 1
test_df['TrainFlag'] = 0
test_df['Survived'] = np.nan

combined = pd.concat([train_df, test_df], sort=False)

# Fill missing values
combined['Age'].fillna(combined['Age'].median(), inplace=True)
combined['Fare'].fillna(combined['Fare'].median(), inplace=True)
combined['Embarked'].fillna(combined['Embarked'].mode()[0], inplace=True)

# Encode categorical features
combined['Sex'] = combined['Sex'].map({'male': 0, 'female': 1})
combined['Embarked'] = combined['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Define features
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
train_data = combined[combined['TrainFlag'] == 1]
test_data = combined[combined['TrainFlag'] == 0]

X = train_data[features]
y = train_data['Survived'].astype(int)
X_test_final = test_data[features]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test_final)

# Split a validation set for visualization
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Train Random Forest
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model.fit(X_train, y_train)
val_preds = model.predict(X_val)

# Evaluate on validation set
print("Validation Accuracy:", accuracy_score(y_val, val_preds))
print("\nClassification Report:\n", classification_report(y_val, val_preds))

# Confusion matrix
conf_matrix = confusion_matrix(y_val, val_preds)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Cross-validation accuracy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X_scaled, y, cv=cv, scoring='accuracy')

# Print scores
print("Cross-validation scores:", cv_scores)
print("Average CV Accuracy:", np.mean(cv_scores))

# Plot CV scores
plt.figure(figsize=(8, 4))
plt.plot(range(1, 6), cv_scores, marker='o', linestyle='--', color='b')
plt.title("Cross-Validation Accuracy per Fold")
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.ylim(0.7, 1.0)
plt.grid(True)
plt.show()

# Final predictions for submission
model.fit(X_scaled, y)
predictions = model.predict(X_test_scaled)

submission = pd.DataFrame({
    'PassengerId': passenger_ids,
    'Survived': predictions.astype(int)
})

submission.to_csv("titanic_submission_rf.csv", index=False)
print("Submission file created: titanic_submission_rf.csv")
```
